```python
import json  
import pandas as pd
f = open('./vinmec/thongtin_tungbenh/cuthe.json', encoding='UTF-8')
data = json.load(f)

list_benh = []

list_vande_tongquat = ['Tổng quan',
 'Nguyên nhân',
 'Triệu chứng',
 'Đường lây truyền',
 'Đối tượng nguy cơ',
 'Phòng ngừa',
 'Các biện pháp chẩn đoán',
 'Các biện pháp điều trị']

for benh in data[:]:
    benh_etl = {} 
    list_infor = benh["cac-van-de-lien-quan"]
 
    ten_benh = str(list_infor[0]['van-de']).replace("  "," ").replace("Hình ảnh bệnh","").replace("Hình ảnh ","").replace("Tổng quan bệnh ","")

    # print(ten_benh)
    vande_list = [] 

    benh_etl['Tên bệnh'] = ten_benh

    for vande in list_infor[1:]:
        vande_list.append(vande["van-de"])

        string_remove_space = ' '.join(str(vande["van-de"]).split())

        ten_vande =  string_remove_space.replace(str(ten_benh),"").replace("bệnh","").strip()
        if ten_vande in list_vande_tongquat:
            benh_etl[ten_vande] = vande["tra-loi"]
            # print("=========",ten_vande,"===", vande["van-de"], "===",ten_benh)
        else:
            # print(ten_benh)
            print(ten_vande,"===", vande["van-de"], "===",ten_benh, "====",string_remove_space)

    # benh_etl['sl'] = len(vande_list)
    # benh_etl["vde"] = vande_list
    list_benh.append(benh_etl)

df = pd.DataFrame(list_benh, columns=['Tên bệnh', 'Tổng quan',
 'Nguyên nhân',
 'Triệu chứng',
 'Đường lây truyền',
 'Đối tượng nguy cơ',
 'Phòng ngừa',
 'Các biện pháp chẩn đoán',
 'Các biện pháp điều trị'])
    
```
