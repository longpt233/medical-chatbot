```python
import json  
import pandas as pd
f = open('./vinmec/thongtin_tungbenh/cuthe.json')
data = json.load(f)

list_benh = []

list_vande_tongquat = ['Tổng quan',
 'Nguyên nhân',
 'Triệu chứng',
 'Đường lây truyền',
 'Đối tượng nguy cơ',
 'Phòng ngừa',
 'Các biện pháp chẩn đoán',
 'Các biện pháp điều trị']

for benh in data[:]:
    benh_etl = {} 
    list_infor = benh["cac-van-de-lien-quan"]
 
    benh_etl['Tên bệnh']= str(list_infor[0]['van-de']).replace("  "," ").replace("Hình ảnh bệnh ","")

    vande_list = [] 

    for vande in list_infor[1:]:
        vande_list.append(vande["van-de"])

        string_remove_space = ' '.join(str(vande["van-de"]).split())

        ten_vande =  string_remove_space.replace(benh_etl['Tên bệnh'],"").replace("bệnh","").strip()
        if ten_vande in list_vande_tongquat:
            benh_etl[ten_vande] = vande["tra-loi"]

        else:
            print(ten_vande,"===", vande["van-de"], "===",benh_etl['Tên bệnh'])

    benh_etl['sl'] = len(vande_list)
    benh_etl["vde"] = vande_list
    list_benh.append(benh_etl)

df = pd.DataFrame(list_benh, columns=['Tên bệnh','sl','vde', 'Tổng quan',
 'Nguyên nhân',
 'Triệu chứng',
 'Đường lây truyền',
 'Đối tượng nguy cơ',
 'Phòng ngừa',
 'Các biện pháp chẩn đoán',
 'Các biện pháp điều trị'])
    
```
